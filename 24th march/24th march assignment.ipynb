# Q2. How did you handle missing data in the wine quality data set during the feature engineering process? Discuss the advantages and disadvantages of different imputation techniques.

---> Missing data handeled using Mean Imputation, Median Imputation and Mode Imputation.


1) Mean Imputation:

---> Mean imputation is easy to implement and does not change the distribution of the data. However, it can lead to biased estimates if there are many missing values.


2) Median Imputation:

---> Median imputation is also easy to implement and is more robust to outliers than mean imputation. However, it can also lead to biased estimates if there are many missing values.


3) Mode Imputation:

---> Mode imputation is useful for categorical data but can lead to biased estimates if there are many missing values.
